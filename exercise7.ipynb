```python
# 7a

from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import numpy as np

gamma = 0.0012
C = 0.85

# Set random seed for reproducibility
seed = 42
np.random.seed(seed)

# Load digits dataset
digits = load_digits()
X, y = digits.data, digits.target

# Train-test split (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=seed)

# Fit
svc = SVC(kernel='rbf', gamma=gamma, C=C)
svc.fit(X_train, y_train)

# Predict
predictions = svc.predict(X_test)
acc = (predictions == y_test).sum() / len(X_test)

np.round(acc, 2)
```

```python
# 7e

from sklearn.datasets import load_digits
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.svm import SVC
import numpy as np

gamma = 0.0012
C = 0.85

# Set random seed for reproducibility
seed = 42
np.random.seed(seed)

# Load digits dataset
digits = load_digits()
X, y = digits.data, digits.target

# Train-test split (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=seed)

param_grid = {
    'C': [0.6, 0.8, 1, 2, 3, 4, 6], 
    'gamma': [0.0001, 0.0006, 0.001, 0.006], 
    'kernel': ['rbf']
} 

svc = SVC(kernel='rbf', gamma=gamma, C=C)
grid = GridSearchCV(svc, param_grid, refit = True, cv = 5, scoring='accuracy') 
grid.fit(X_train, y_train)

grid.best_params_, np.round(grid.score(X_test, y_test), 2)
```
